```python
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, ne_chunk
from nltk.tree import Tree
print("Libraries imported successfully!")
```

```python
sample_text = """
Natural Language Processing (NLP) is a subfield of Artificial Intelligence.
It helps computers understand, interpret, and generate human language.
Barack Obama was the 44th President of the United States.
The quick brown fox jumps over the lazy dog!
Running, runs, and runner are all forms of the word 'run'.
"""
print("Sample text loaded!")
print(sample_text)
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

cleaned_text = clean_text(sample_text)
print("=== Cleaned Text ===")
print(cleaned_text)
```

```python
tokens = word_tokenize(cleaned_text)
print("=== Tokens ===")
print(tokens)
print(f"\nTotal tokens: {len(tokens)}")
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]
print("=== After Stopword Removal ===")
print(filtered_tokens)
print(f"\nTokens removed: {len(tokens) - len(filtered_tokens)}")
```

```python
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print("=== Stemmed Tokens ===")
print(stemmed_tokens)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(" Lemmatized Tokens ")
print(lemmatized_tokens)
```

```python
original_tokens = word_tokenize(cleaned_text)
pos_tags = pos_tag(original_tokens)
print("=== POS Tags ===")
for word, tag in pos_tags[:20]:
    print(f"  {word:20} → {tag}")
```

```python
ner_text = "Barack Obama was the 44th President of the United States."
ner_tokens = word_tokenize(ner_text)
ner_pos = pos_tag(ner_tokens)
ner_tree = ne_chunk(ner_pos)

print("=== Named Entities Found ===")
for subtree in ner_tree:
    if isinstance(subtree, Tree):
        entity = " ".join([word for word, tag in subtree.leaves()])
        label = subtree.label()
        print(f"  Entity: {entity:30} → Type: {label}")
```

```python
print("=" * 55)
print("        NLP PREPROCESSING ENGINE - SUMMARY")
print("=" * 55)
print(f"Original text length : {len(sample_text.split())} words")
print(f"After tokenization   : {len(tokens)} tokens")
print(f"After stopword removal: {len(filtered_tokens)} tokens")
print(f"Sample stemmed words : {stemmed_tokens[:5]}")
print(f"Sample lemmatized    : {lemmatized_tokens[:5]}")
print(f"Total POS tags       : {len(pos_tags)}")
print("=" * 55)
print("NLP Preprocessing Engine built successfully!")
```
